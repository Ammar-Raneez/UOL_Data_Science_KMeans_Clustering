```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from sklearn.cluster import KMeans

data = pd.read_csv('Banknote_authentication.csv')

mean_coord = data.mean(0)
sd = data.std(0)

# print(sd[0], sd[1])
# print(mean_coord[0], mean_coord[1])

#fig, graph = plt.subplots()
#ellipse = patches.Ellipse([mean_coord[0], mean_coord[1]], sd[0]*2, sd[1]*2, alpha=0.3)

km_data = np.column_stack((data['V1'], data['V2']))
km_result = KMeans(n_clusters=2).fit(km_data)
clusters = km_result.cluster_centers_

plt.scatter(data['V1'], data['V2'], alpha=0.5)
plt.scatter(clusters[:,0], clusters[:,1], s=1000)
#plt.scatter(mean_coord[0], mean_coord[1], s=200)

plt.xlabel('V1')
plt.ylabel('V2')

#graph.add_patch(ellipse)
plt.show()
# plt.savefig("Figure")
```
